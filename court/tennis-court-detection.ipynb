```python
from ultralytics import YOLO

def train_phase_1():
    model = YOLO("yolo11s.pt") 

    model.train(
        data="/kaggle/input/tennis-dataset/data/data.yaml",  
        
        epochs=20,          
        batch=32,          
        imgsz=960,
        
        workers=0,          
        device=[0, 1],       # Use BOTH GPUs
        
        optimizer='AdamW',   
        lr0=0.001,           
        fliplr=0.5,          
        close_mosaic=10,   
        
        project="/kaggle/working/runs", 
        name="court_detector_phase1",
        exist_ok=True,
    )

if __name__ == '__main__':
    train_phase_1()
```
